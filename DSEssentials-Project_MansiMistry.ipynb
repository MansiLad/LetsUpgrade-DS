22. Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot it on a histogram and comment on its distribution.

```python
#data for the graph
hist_data = []
for i in range(0,5000,50):
    newdata = pubgdata.iloc[i:50+i]
    mean = newdata['damageDealt'].mean()
    hist_data.append(mean)
    
hist_data = np.array(hist_data)
```

```python
sns.histplot(data=hist_data, bins = 50)
```
